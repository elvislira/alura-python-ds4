# Aula 04 - Manipulando os dados

* Importando os dados novamente

```python
dados = pd.read_csv('dados/aluguel.csv', sep=';')
dados
```

* Criando colunas numéricas

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']
dados.head()
```

* Criando coluna descritiva

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' \
    + dados['Quartos'].astype(str) + ' quarto(s) e ' + dados['Vagas'].astype(str) \
    + ' vaga(s) de garagem.'
dados.head()
```

* Criando coluna categórica

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: 'Sim' if x > 0 else 'Não')
dados.head()
```

* Salvando os dados para o time de desenvolvimento

```python
dados.to_csv('dados_tratados/dados_completos_dev.csv', index=False, sep=';')
```
